```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x=np.linspace(-3,3,100)
noise=np.random.normal(0,1,100)
y=0.5*x**2+2*x+2+noise
# print(x)
# print(y)
plt.scatter(x,y,c='b')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x.reshape(-1,1),y)
predicted=model.predict(x.reshape(-1,1))
plt.scatter(x,y,c='b')
plt.plot(x,predicted,c='r')
plt.show()
```

```python
x_in=x.reshape(-1,1)
x2=np.hstack([x_in,x_in**2])
# print(x)
# print(x_in)
# print(x2)
model = LinearRegression()
model.fit(x2,y)
predicted=model.predict(x2)
plt.scatter(x,y,c='b')
plt.plot(x,predicted,c='r')
plt.show()
print(model.coef_,model.intercept_)
```

```python
#
from sklearn.preprocessing import PolynomialFeatures
poly= PolynomialFeatures(degree=2)#2
poly.fit(x_in)
x2=poly.transform(x_in)
print(x2.shape)#3

model = LinearRegression()
model.fit(x2,y)
predicted=model.predict(x2)
plt.scatter(x,y,c='b')
plt.plot(x,predicted,c='r')
plt.show()
print(model.coef_,model.intercept_)
```

```python
#
from sklearn.preprocessing import PolynomialFeatures
poly= PolynomialFeatures(degree=3)#3
poly.fit(x_in)
x2=poly.transform(x_in)
print(x2.shape)#3

model = LinearRegression()
model.fit(x2,y)
predicted=model.predict(x2)
plt.scatter(x,y,c='b')
plt.plot(x,predicted,c='r')
plt.show()
print(model.coef_,model.intercept_)
```

```python
from sklearn.metrics import mean_squared_error
print('',mean_squared_error(y,predicted))
```
